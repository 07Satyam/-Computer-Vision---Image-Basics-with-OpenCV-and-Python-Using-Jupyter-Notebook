# OpenCV

### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

### Import OpenCV

```python
import cv2
```

Get the image with the imread

```python
img=cv2.imread('Downloads/AlluPic.jpg')
```

### Always check the type

```python
type(img)
```

### Get the image shape

```python
img.shape
```

Let's see the image with the imshow

```python
plt.imshow(img)
```

```python
img_fix=cv2.cvtColor(img,
                     cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

### Scale it to Gray and check the Shape

```python
img_gray=cv2.imread('Downloads/AlluPic.jpg',
                    cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

### Resize the image

```python
img_new=cv2.resize(img_fix,
                  (200,400))
```

```python
plt.imshow(img_new)
```

### Resize with Ratio

```python
width_ratio=0.5
height_ratio=0.5
```

```python
img2=cv2.resize(img_fix,
               (0,0),
               img_fix,
              width_ratio,
              height_ratio)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

### Flip on Horizontal Axis

```python
img_3=cv2.flip(img_fix,0)
plt.imshow(img_3)
```

### Flip on Vertical Axis

```python
img_3=cv2.flip(img_fix,1)
plt.imshow(img_3)
```

### Flip on Horizontal and on Vertical Axis

```python
img_3=cv2.flip(img_fix,-1)
plt.imshow(img_3)
```

### Change the size of our canvas

```python
last_img=plt.figure(figsize=(10,7))
ilp=last_img.add_subplot(111)
ilp.imshow(img_fix)
```

# 02_Draw_shapes_on_image.

### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline 
```

### Create a black image to work

```python
black_img=np.zeros(shape=(512,512,3),
                  dtype=np.int16) 
```

### Get the Shape of the image

```python
black_img.shape
```

### Show it

```python
plt.imshow(black_img)
```

### Draw a Circle

The center is the first number on x-axis and second on y-axis

```python
cv2.circle(img=black_img,
          center=(400,100),
          radius=50,
          color=(255,0,0),
          thickness=8)
```

```python
plt.imshow(black_img)
```

### Filled Circle

```python
cv2.circle(img=black_img,
          center=(400,200),
          radius=50,
          color=(0,255,0),
          thickness=-1)

plt.imshow(black_img)
```

### Draw a Rectangle

```python
cv2.rectangle(black_img,
          pt1=(200,200),
          pt2=(300,300),
          color=(0,255,0),
          thickness=5)

plt.imshow(black_img)
```

### Draw Triangle

```python
vertices=np.array([[10,450],
                  [110,350],
                  [180,450]],
                 np.int32)

pts=vertices.reshape(-1,1,2)

cv2.polylines(black_img,
             [pts],
             isClosed=True,
             color=(0,0,255),
             thickness=3)

plt.imshow(black_img)
```

### Filled Rectangle

```python
cv2.rectangle(black_img,
          pt1=(200,50),
          pt2=(300,150),
          color=(137,79,213),
          thickness=-1)

plt.imshow(black_img)
```

### Filled Triangle

```python
vertices=np.array([[10,250],
                  [110,150],
                  [180,250]],
                 np.int32)

pts=vertices.reshape(-1,1,2)

cv2.fillPoly(black_img,
             [pts],
             color=(255,167,201))

plt.imshow(black_img)
```

### Draw Line

```python
cv2.line(black_img,
          pt1=(512,0),
          pt2=(0,512),
          color=(255,0,255),
          thickness=3)

plt.imshow(black_img)
```

### Write Text 

```python
font=cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(black_img,
           text='Rhyme',
           org=(210,500),
           fontFace=font,
           fontScale=3,
           color=(255,255,0),
           thickness=3,
           lineType=cv2.LINE_AA)

plt.imshow(black_img)
```
